### Метод наименьших квадратов

Будем подгонять полином методом наименьших квадратов:
$$X = \begin{pmatrix}
 1 & x_1 & \dots & x_1^K \\
 & & \dots & \\
 1 & x_N & \dots & x_N^K 
\end{pmatrix}$$
$$y = Xw$$
$$R(w) = \|Xw - y\|^2 = \sum_{n=1}^N (w^Tx_n - y_n)^2$$
$$\frac{\partial R}{\partial w} = 2 X^T(Xw - y) = 0$$
$$w_\star = (X^TX)^{-1}X^Ty$$

```python
# fit polynomial of degree K
K = 12
penalty = 10.0
X = np.vander(x, K+1)
w = np.linalg.solve(np.dot(X.T, X) + penalty*np.eye(K+1), np.dot(X.T, y))
w

```

```python
X_plot = np.vander(x_plot, K+1)
y_hat_plot = np.dot(X_plot, w)

# show data and the ground-truth
plot(x_plot, y_hat_plot, 'r')
plot(x_plot, f(x_plot), label="ground truth")
scatter(x, y, label="training points")
```

### Функция

```python
def polynomial_fit(x, y, K, penalty=0.0, x_test=None):
    """Least squares polynomial fit."""
    X = np.vander(x, K+1)
    w = np.linalg.solve(np.dot(X.T, X) + penalty*np.eye(K+1), np.dot(X.T, y))
    if x_test is not None:
        X_test = np.vander(x_test, K+1)
        y_test_hat = np.dot(X_test, w)
        return w, y_test_hat
    return w
```

```python
figure(figsize=(15, 10))
xlim([-2, 12])
ylim([-15, 15])

plot(x_plot, f(x_plot), label="ground truth", linewidth=4)
scatter(x, y, label="training points")

Ks = [2, 5, 12, 15, 20]
for K in Ks:
    w, y_plot = polynomial_fit(x, y, K, x_test=x_plot)
    plot(x_plot, y_plot, label='degree %d' % K, linewidth=1.2)
    
    
legend()
```
